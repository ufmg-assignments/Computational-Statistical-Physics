## Funções

```python
def gerarRede(L):

    N = L*L
    rede = [np.random.choice([1,-1]) for i in range(N)]

    return np.array(rede)
```

```python
def gerarVizinhos(L):

  N = L*L
  vizinhos = np.zeros((N, 4), dtype = np.int32)

  for k in range(N):

      vizinhos[k][0] = k + 1
      if((k+1)%L) == 0:
          vizinhos[k][0] = k + 1 - L

      vizinhos[k][1] = k + L
      if(k > N - 1 - L):
          vizinhos[k][1] = k + L - N
      vizinhos[k][2] = k - 1

      if(k%L == 0):
          vizinhos[k][2] = k + L - 1
          
      vizinhos[k][3] = k - L
      if(k < L):
          vizinhos[k][3] = k + N - L

  return vizinhos
```

```python
def calcularEnergia(s, viz):

    energia = 0
    for i in range(len(s)):
        h = s[viz[i][0]] + s[viz[i][1]]
        energia = energia - s[i]*h

    return energia
```
